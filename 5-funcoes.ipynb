```python
import numpy as np
```

```python
array = np.array([1,2,3,4,5])
array
```

```python
soma = np.add(array,10)
soma
```

```python
subtração = np.subtract(array , 5)
subtração
```

```python
multiplicacao = np.multiply(array,2)
multiplicacao
```

```python
divisao = np.divide(array,2)
divisao
```

```python
#funçoes trigonometricas 
seno = np.sin(array)
seno

cosseno = np.cos(array)
cosseno
```

```python
#funcoes estatristicas 
media = np.mean(array)
media
```

```python
desvio_padrao = np.std(array)
desvio_padrao
```

```python
#comparação
coparacao_maior_que_3 = array > 3
coparacao_maior_que_3
```

```python
#combinando funções
resultado = np.sqrt(array) + np.power(array,2)
resultado
```

```python
#funcoes exponenciais
exponencial = np.exp(array)
exponencial
```

```python
#logaritimo 
logaritimo = np.log(array)
logaritimo
```

```python
#Arredondamento
arredondamento = np.round(array,2)
arredondamento
```

```python
#Potência 
potencia = np.power(array ,3)
potencia
```
