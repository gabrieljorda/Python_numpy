```python
notas = np.array([50,67,72,48,90,85,60,50])
notas
```

```python
#pesqueisar quais alunos foram aprovados (nota >60)
alunos_aprovado = np.where(notas >60)
alunos_aprovado
```

```python
notas[alunos_aprovado]
```

```python
nota_maxima = np.max(notas)
nota_maxima
```

```python
indice_maximo = np.argmax(notas)
indice_maximo
```
